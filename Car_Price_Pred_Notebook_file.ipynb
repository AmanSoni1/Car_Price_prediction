## Cleaning

- Backup created so that I don't load the file again and again

```python
backup = car.copy() 
```

- isnumeric Gives true if numeric else false

```python
car = car[car['year'].str.isnumeric() ]
```

- changing year from obj to int

```python
car['year'] = car['year'].astype(int)
```

- Price's Ask For Price removal

```python
car = car[car['Price'] != 'Ask For Price']
```

- Replace ',' to empty string ' ' and change object to int

```python
car['Price']=car['Price'].str.replace(',','').astype(int)
```

- kms_driver ',' replacement and 'kms' replacement

```python
car['kms_driven']=car['kms_driven'].str.split(' ').str.get(0).str.replace(',','')
```

```python
car = car[car['kms_driven'].str.isnumeric()]
```

```python
car['kms_driven'] = car['kms_driven'].astype(int)
```

- Fuel Type nan removal

```python
car = car[~car['fuel_type'].isna()]
```

- Now Keep only 3 words of Name

```python
car['name'] = car['name'].str.split(' ').str.slice(0,3).str.join(' ')
```

- reset Index 

```python
car = car.reset_index(drop=True)
```

```python
car=car[car['Price']<6e6].reset_index(drop=True)
```

# Model

### Step 1 > Extract labels and Features

```python
X = car.drop(columns = 'Price')
y = car['Price']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```python
ohe = OneHotEncoder()
ohe.fit(X[['name','company','fuel_type']])
```

```python
ohe.categories_
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['name','company' ,'fuel_type']), 
                                       remainder= 'passthrough' )
```

```python
lr = LinearRegression()
```

```python
pipe = make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

### for different set my r2 score varied a lot
 ### So I looped through 1000 sets and chose the best set

```python
scores=[]
for i in range(1000):
    X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state=i)
    lr = LinearRegression()
    pipe=make_pipeline(column_trans,lr)
    pipe.fit(X_train,y_train)
    y_pred=pipe.predict(X_test)
    scores.append(r2_score(y_test,y_pred))
```

```python
np.argmax(scores)
```

```python
scores[np.argmax(scores)]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2, random_state=np.argmax(scores))
lr = LinearRegression()
pipe=make_pipeline(column_trans,lr)
pipe.fit(X_train,y_train)
y_pred=pipe.predict(X_test)
r2_score(y_test,y_pred)
```

```python
pickle.dump(pipe,open('LinearRegressionModel.pkl','wb'))
```
